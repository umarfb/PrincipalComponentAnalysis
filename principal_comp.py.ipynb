```python
import numpy as np
import pandas as pd
```

```python
# Read .csv file containing wavelet coefficients
sn_coeffs = pd.read_csv('sn_wave_coeffs.csv')

sn_coeffs_hdr = list(sn_coeffs)

wv_coeffs = sn_coeffs_hdr[0:-1]
sn_type = sn_coeffs_hdr[-1]

# Seperate out coefficients and type
wv_coeffs_df = sn_coeffs.loc[:,wv_coeffs]
sn_type_df = sn_coeffs.loc[:,sn_type]
```

```python
# Standardize the data, column by column
for col in wv_coeffs_df:
    col_data = wv_coeffs_df[col]
    
    # Get mean and std. deviation
    col_mean = np.mean(col_data)
    col_stdv = np.std(col_data)
    
    # Standardise array
    wv_coeffs_df[col] = col_data - col_mean
    wv_coeffs_df[col] = col_data / col_stdv
    
wv_coeffs_df
```

```python
# Attempt principal component analysis
from sklearn.decomposition import PCA

pca = PCA(n_components = 20)

principal_comps = pca.fit_transform(wv_coeffs_df)

principalDf = pd.DataFrame(data = principal_comps)
principalDf['sn_type'] = sn_type_df

# Remove SN with unknown type
principalDf = principalDf.drop(142)
principalDf = principalDf.drop(3)
principalDf
```

```python
sn_type_arr = principalDf['sn_type'].values

for i, type_val in enumerate(principalDf['sn_type']):
    if 'Ia' in type_val and '/c' not in type_val:
        sn_type_arr[i] = 'Ia'
    if 'Ib' in type_val and 'II' not in type_val:
        sn_type_arr[i] = 'Ib'
    if 'Ic' in type_val:
        sn_type_arr[i] = 'Ic'
    if type_val == 'II P':
        sn_type_arr[i] = 'II P'
    if 'IIn' in type_val:
        sn_type_arr[i] = 'IIn'
    if 'SLSN' in type_val:
        sn_type_arr[i] = 'SLSN'
    if 'Ca-rich' in type_val:
        print(type_val)
    
principalDf['sn_type'] = sn_type_arr
principalDf['sn_type']
```

```python
sn_type = principalDf['sn_type'].values

# Get unique sn types
sn_classes = set(sn_type)
sn_type_list = []
for u_class in sn_classes:
    sn_type_list.append(u_class)
    
sn_type_list = sorted(sn_type_list)
# Break dataframe by sn types
type_dfs = []
for target_type in sn_type_list:
    sing_type_df = principalDf.loc[principalDf['sn_type'] == target_type]
    type_dfs.append(sing_type_df)

type_dfs[0]
```
